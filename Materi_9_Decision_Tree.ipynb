```python
from sklearn import tree
```

```python
# Database: Gerbang Logika AND
# x = Data,y = Target
```

```python
x = [[0,0],[0,1],[1,0],[1,1]]
y = [0, 0, 0, 1]
```

```python
#Training and Clasify
clf = tree.DecisionTreeClassifier()
clf = clf.fit(x,y)
```

```python
#Prediction
print("Logika AND Metode Decision Tree")
print("10 10 5 = ",clf.predict([[0,0]]))
print("0 1 = ",clf.predict([[0,1]]))
print("1 0 = ",clf.predict([[1,0]]))
print("1 1 = ",clf.predict([[1,1]]))
```

```python
from google.colab import drive
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt
```

```python
# Mount Google Drive
drive.mount('/content/drive')

#Path ke file di Google Drive
FileDB ='/content/drive/My Drive/Sinus[1].txt' #Sesuaikan path file
Database = pd.read_csv(FileDB,sep=",", header=0)

#Lihat data
print("------------------")
```

```python
# x data y target
x = Database[['Feature']] #replace with your actual column names
y = Database.Target
```

```python
reg = DecisionTreeRegressor(random_state=1)
reg = reg.fit(x,y)
```

```python
# Display preticted data
xx = np.arange(1, 21, 1)
n = len(xx)
print("xx(i) Decision Tree")
for i in range(n):
   y_dct = reg.predict([[xx[i]]])
   print('{:.2f}'.format(xx[i]),y_dct)

#Plot the preticted data
y_dct2 = reg.predict(x)
plt.figure()
plt.plot(x,y_dct2, color='red')
plt.scatter(x,y, color='blue')
plt.title('Prediksi Data Menggunakan Decision Tree')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['Decision Tree','Data'], loc=2)
plt.show()
```
